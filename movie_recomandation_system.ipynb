```python
# Install Kaggle
!pip -q install kaggle

# Upload kaggle.json


from google.colab import files
files.upload()

# Setup Kaggle API


import os

os.makedirs("/root/.kaggle", exist_ok=True)

!cp kaggle.json /root/.kaggle/
!chmod 600 /root/.kaggle/kaggle.json


# Download Dataset


!kaggle datasets download -d tmdb/tmdb-movie-metadata

!unzip -o tmdb-movie-metadata.zip

# Import Libraries

import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load Dataset

movies = pd.read_csv("tmdb_5000_movies.csv")

print("Dataset Loaded Successfully!")

print("\nShape:")
print(movies.shape)

print("\nFirst Five Movies:")
print(movies[['title','genres']].head())

# Data Cleaning


movies = movies[['title','overview']]

movies['overview'] = movies['overview'].fillna('')

# TF-IDF Vectorization

tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(movies['overview'])

print("\nTF-IDF Matrix Created!")


# Cosine Similarity


similarity = cosine_similarity(tfidf_matrix)

print("Similarity Matrix Created!")

# Recommendation Function

def recommend(movie_name):

    movie_name = movie_name.lower()

    movie_titles = movies['title'].str.lower()

    if movie_name not in movie_titles.values:
        print("\nMovie Not Found!")
        return

    index = movie_titles[movie_titles == movie_name].index[0]

    scores = list(enumerate(similarity[index]))

    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print("\nTop 10 Recommended Movies:\n")

    count = 0

    for i in scores[1:11]:

        print(movies.iloc[i[0]].title)

        count += 1


# User Input


movie = input("Enter Movie Name: ")

recommend(movie)

print("\nRecommendation Completed Successfully!")
```
