```python
# ==========================================
# Rule-Based AI Chatbot
# Internship Project 1 - DecodeLabs
# ==========================================

# Display welcome message
print("===================================")
print("     Welcome to AI Chatbot 🤖")
print("Type 'bye' or 'exit' to quit.")
print("===================================")

# Continuous loop
while True:

    # Take input from user
    user_input = input("\nYou: ")

    # Convert input to lowercase
    user_input = user_input.lower()

    # Exit conditions
    if user_input == "bye" or user_input == "exit" or user_input == "quit":
        print("Bot: Goodbye! Have a nice day 😊")
        break

    # Greetings
    elif user_input == "hello" or user_input == "hi" or user_input == "hey":
        print("Bot: Hello! How can I help you?")

    # Asking name
    elif "your name" in user_input:
        print("Bot: My name is RuleBot, your AI assistant.")

    # Asking about AI
    elif "what is ai" in user_input:
        print("Bot: AI stands for Artificial Intelligence. It enables machines to perform tasks that normally require human intelligence.")

    # Asking about Python
    elif "python" in user_input:
        print("Bot: Python is a popular programming language used in AI, Machine Learning, and Web Development.")

    # Asking about internship
    elif "internship" in user_input:
        print("Bot: This internship helps you learn AI concepts through practical projects.")

    # Asking about chatbot
    elif "chatbot" in user_input:
        print("Bot: A chatbot is a program that communicates with users through text or voice.")

    # Asking time
    elif "time" in user_input:
        from datetime import datetime
        current_time = datetime.now().strftime("%I:%M:%S %p")
        print("Bot: Current time is", current_time)

    # Asking date
    elif "date" in user_input:
        from datetime import datetime
        current_date = datetime.now().strftime("%d-%m-%Y")
        print("Bot: Today's date is", current_date)

    # Asking how are you
    elif "how are you" in user_input:
        print("Bot: I am fine and ready to help you!")

    # Thank you
    elif "thank you" in user_input or "thanks" in user_input:
        print("Bot: You're welcome! 😊")

    # Default response
    else:
        print("Bot: Sorry, I don't understand that. Please try another question.")
```
